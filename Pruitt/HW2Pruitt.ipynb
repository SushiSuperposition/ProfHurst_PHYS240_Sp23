```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sumUp(N):
    """Takes input of N, returns sum of 1/n from n = 1 to n = N"""
    S = 0
    i = 1
    while i <= N:
        S += 1/i
        i += 1
        
    return(S)

def sumDo(N):
    """Takes input of N, returns sum of 1/n from n = N to n = 1"""
    S = 0
    while N >= 1:
        S += 1/N
        N -= 1
        
    return(S)
```

```python
nVals = np.arange(1, 1000, 1) # array of n values from 1 to 999, with stepsize 1

sup = np.array([sumUp(n) for n in nVals]) # upward sum value array using list comprehension
sdo = np.array([sumDo(n) for n in nVals]) # downward sum value array using list comprehension


top = sup - sdo
bot = np.abs(sup) + np.abs(sdo)
```

```python
plt.figure()
plt.loglog(nVals, top/bot, 'k.')
plt.title('log-log plot to observe dynamic range')
plt.xlabel('N')
plt.ylabel('$\\frac{S^(up) - S^(down)}{|S^(up)| + |S^(down)|}$')
plt.show()
```

```python
plt.figure()
plt.plot(nVals, top/bot, 'k.')
plt.title('linear regime')
plt.xlabel('N')
plt.ylabel('$\\frac{S^(up) - S^(down)}{|S^(up)| + |S^(down)|}$')
plt.show()
```

##### In general, the downwards sum is more precise as the accumulation of round-off errors happens as the upwards sum runs from lower values to higher, where memory is not allocated correctly for successive terms that require more precision. The downwards sum avoids this by starting with values that already require higher precision and thus avoids rounding more often.
